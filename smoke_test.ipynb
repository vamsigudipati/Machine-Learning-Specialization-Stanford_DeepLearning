# Machine Learning Specialization — Environment Smoke Test

Verifies that all libraries required across the three courses are correctly installed and importable.

| Course | Key Libraries |
|--------|--------------|
| Supervised ML | `numpy`, `matplotlib`, `scikit-learn` |
| Advanced Learning Algorithms | `tensorflow`, `keras`, `numpy`, `matplotlib` |
| Unsupervised / Recommenders / RL | `tensorflow`, `numpy`, `pandas`, `matplotlib` |

```python
import sys

results = []

# ── Core libraries ────────────────────────────────────────────────
libs = [
    "numpy",
    "pandas",
    "matplotlib",
    "sklearn",       # scikit-learn
    "tensorflow",
    "keras",
]

for lib in libs:
    try:
        mod = __import__(lib)
        version = getattr(mod, "__version__", "n/a")
        results.append((lib, "✅", version))
    except ImportError:
        results.append((lib, "❌", "NOT FOUND"))

# ── Report ────────────────────────────────────────────────────────
print(f"Python : {sys.version}\n")
print(f"{'Library':<15} {'Status':<6} {'Version'}")
print("-" * 40)
for name, status, version in results:
    print(f"{name:<15} {status:<6} {version}")

all_ok = all(s == "✅" for _, s, _ in results)
print("\n" + ("✅ Environment ready — all dependencies satisfied." if all_ok
              else "⚠️  One or more dependencies are missing. Install them with pip."))
```
